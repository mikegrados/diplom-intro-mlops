# 0. Introduccion a POO

La programación orientada a objetos se basa en el concepto de crear un modelo del problema de destino en sus programas. La programación orientada a objetos disminuye los errores y promociona la reutilización del código. 

Python es un lenguaje de programación orientado a objetos.
Casi todo en Python es un objeto, con sus propiedades y métodos.
Una clase es como un constructor de objetos o un "modelo" para crear objetos.

## 0.1  ¿Por qué usar POO?
Muchas cosas se pueden lograr hacer con Python mientras el codigo sea simple y facil de entender. Pero conforme los proyectos se van volviendo mas grandes y complejos, sera necesario encontrar una forma de mantener los programas y la informacion organizados, estructurados, faciles de modificar en cuanto haya nuevos integrantes en el equipo de trabajo, o se vayan agregando nuevas características.

Es aqui donde POO puede ayudar bastante a crear programas modulares, faciles de planificar, mantener y aprender por nuevos programadores. 



# 1 - Objetos en Python

## 1.1 - Objetos
> Los objetos pueden no ser cosas tangibles que puedas captar o sentir, mas bien, son modelos de algo que puede hacer ciertas cosas y que tambien están sujetos a que se les hagan ciertas cosas. Formalmente, **un objeto es una colección de datos y comportamientos asociados**.

> Cada objeto, de igual modo que la entidad de la vida real a la que representa, tiene un estado (es decir, unos atributos con unos valores concretos) y un comportamiento (es decir, tiene funcionalidades o sabe hacer unas acciones concretas).

**Ejemplo de objeto: un auto.**

> La clase se representa por una plantilla o esqueleto a partir del cual se creará el auto.

> Cuando se fabrica el auto, se le considera como un objeto, totalmente diferenciado del pavimento en donde está estacionado.

> Sus **atributos/propiedades** son su color, marca, modelo, número de matrícula, número de puertas, etc.

> Los **métodos** son todo aquello que nos ofrece el auto, como hacer sonar una bocina cuando tocamos el claxon, llevarnos por la carretera, reducir la velocidad al pisar el freno, etc.

Definir clases permite trabajar con código reutilizable. Puesto que desde una clase se puede crear una instancia y así reutilizar el código escrito para esta si tener que volver a escribir el código para la instancia. La instancia toma el patrón de la clase padre. Sin embargo, las variables son idependientes.

# 3 - `self` and `__init__()`

## 3.1 - `self`
El parámetro `self` es una referencia a la instancia actual de la clase y se utiliza para acceder a las variables que pertenecen a la clase.

No tiene que llamarse `self`, puedes llamarlo como quieras, pero tiene que ser el primer parámetro de cualquier método en la clase:

## 3.2 - `__init__()`

Para comprender el significado de las clases en Python, debemos entender primero el método incorporada `__init__()` .

Todas las clases tienen un método llamad `__init__()`, que siempre se ejecuta cuando se inicia la clase.

Se usa el método `__init__()` para asignar valores a las propiedades del objeto u otras operaciones que sean necesarias cuando se crea el mismo.

```python
class Persona:
  def __init__(miobjeto, nombre, edad): # 1)
    miobjeto.nombre = nombre
    miobjeto.edad = edad

  def mi_metodo(abc): # 2)
    print("Hola, mi nombre es " + abc.nombre + " y mi edad es " + str(abc.edad))

p1 = Persona("Carlos", 26) # 3)
print(p1.nombre)
print(p1.edad)
p1.mi_metodo() # 4)
```

En el codigo anterior:
* 1) Usamos la palabra `miobjeto` en lugar de `self`.
* 2) Usamos la palabra `abc`  en lugar de `self`.
* 3) Creamos una instancia de la clase Persona y le pasamos como argumentos el nombre "Carlos" y la edad 26.
* 4) Llamamos al método definido `mi_metodo()` que imprime el nombre y la edad.

**Nota**

En la definición de un método los valores que se reciben se denominan *parámetros*, pero durante la llamada los valores que se envían se denominan *argumentos*.

##3.3 - Ejemplos

# 4 - Herencia
Es la capacidad de una clase de heredar atributos y métodos de otra, además de agregar nuevos y/o modificar los existentes.

## 4.1 Herencia simple
Revise las clases debajo y la forma de instanciarlas, pasando argumentos para crear objetos.

```python
class Libro:
  def __init__(self, titulo, precio, autor, paginas):
      self.titulo = titulo
      self.precio = precio
      self.autor = autor
      self.paginas = paginas

class Revista:
  def __init__(self, titulo, precio, periodo, editor):
      self.titulo = titulo
      self.precio = precio
      self.periodo = periodo
      self.editor = editor

class Periodico:
  def __init__(self, titulo, precio, periodo, editor):
      self.titulo = titulo
      self.precio = precio
      self.periodo = periodo
      self.editor = editor

l1 = Libro(titulo="100 años de soledad", autor="Gabriel Garcia Marquez", paginas=300, precio=300.0)
r1 = Revista(titulo="Ciencias", editor="UNAM", precio=0, periodo="Mensual")
p1 = Periodico("El Financiero.", "Grupo Multimedia Lauman", 3.0, "Diario")

print(l1.autor)
print(p1.editor)
print(l1.precio, p1.precio, r1.precio)
```

```python
print(r1.titulo)
print(r1.precio)
print(r1.periodo)
print(r1.editor)
```

Aunque el codigo funciona muy bien, hay atributos en las clases que estan duplicados. Las tres clases tienen los atributos `titulo` y `precio`. Mientras que la clase `Revista` y `Periodico` tienen los atributos `periodo` y `editor`.

Para organizar el codigo de estas clases, y permitir que nuevas puedan ser agregadas en el futuro, vamos a ordenar el codigo. El resultado se refleja en la celda de abajo.

```python
class Publicacion:
  def __init__(self, titulo, precio):
      self.titulo = titulo
      self.precio = precio

class Libro(Publicacion):
  def __init__(self, titulo, precio, autor, paginas):
    super().__init__(titulo, precio)
    self.autor = autor
    self.paginas = paginas

class Periodico(Publicacion):
  def __init__(self, titulo, precio, periodo, editor):
      super().__init__(titulo, precio)
      self.periodo = periodo
      self.editor = editor

class Revista(Periodico):
  def __init__(self, titulo, precio, periodo, editor):
      super().__init__(titulo, precio, periodo, editor)

```

```python
libro = Libro("Sapiens", 500, "Yuval Noah", 443)
print(libro.precio, libro.titulo, libro.autor, libro.paginas)
```

```python
periodico = Periodico("El Financiero", 10, "Diario", "Grupo Multimedia Lauman")
print(periodico.titulo, periodico.precio, periodico.periodo, periodico.editor)
```

```python
revista = Revista("Ciencias", 0, "Mensual", "UNAM")
print(revista.titulo, revista.precio, revista.periodo, revista.editor)
```

`super()` es un método en Python que nos permite llamar a un método en una clase principal. Esto puede ser útil para la herencia múltiple, donde puede usar super para llamar a un método en una clase principal que no es el último elemento en el orden de resolución del método.


## 4.2 Herencia multiple
La herencia múltiple permite heredar métodos y atributos de dos o más clases base.


A continuacion se observa un ejemplo de herencia múltiple.

```python
class A:
  def __init__(self):
    super().__init__()
    self.foo = "foo"
    self.name = "Clase A"
  
  def mensaje(self):
    print("En Clase A")

class B:
  def __init__(self):
    super().__init__()
    self.bar = "bar"
    self.name = "Clase B"

  def mensaje(self):
    print("En Clase B")


class C(A, B):
  def __init__(self):
    super().__init__()

  def show_attribs(self):
    print(self.foo)
    print(self.bar)
    print(self.name)

  def mensaje(self): # Anulación de metodo de clase base
    print("En Clase C")


c = C()
print(C.__mro__)
c.show_attribs()
c.mensaje()
```

**Notas** 

El MRO es el orden en que se busca un método en la jerarquía de clases. Es una secuencia que usa Python para encontrar un método en una clase. Es principalmente útil en el caso de la herencia para buscar un método en las clases principales.

La clase `object` es el ancestro de cada clase en Python. Si hace `object.__bases__`, obtendrá un valor vacío.

En Python 2.x, todavía es obligatorio heredar explícitamente objetos como la clase `BaseClass(object)`, pero desde Python 3.x, ya no es necesario. Es generado por default.

# 5 - Polimorfismo

El polimorfismo es la técnica que nos posibilita que al invocar un determinado método de un objeto, podrán obtenerse distintos resultados según la clase del objeto. Esto se debe a que distintos objetos pueden tener un método con un mismo nombre, pero que realice distintas operaciones.

```python
print(len("polimorfismo"))
print(len([1,2,3]))
```

```python
class Gato:
	def hablar(self):
		print("MIAU")	

class Perro:
	def hablar(self):
		print("GUAU")	

def escuchar_mascota(animal):
	animal.hablar()	
 
gato = Gato()
perro = Perro()

escuchar_mascota(gato)
escuchar_mascota(perro)
```

Podemos llamar a la función `escuchar_mascota()` pasando como algo argumento, en este caso el animal. Luego, a ese algo se la llama su función `hablar()`, aún cuando esta venga de diferentes clases. Esto es **polimorfismo**.

## 5.1 Polimorfismo y Herencia
El polimorfismo nos permite definir métodos en la clase secundaria que tienen el mismo nombre que los métodos en la clase principal. 


En la herencia, la clase secundaria hereda los métodos de la clase principal. Sin embargo, **es posible modificar un método en una clase secundaria que haya heredado de la clase principal**. 


Esto es particularmente útil en los casos en que el método heredado de la clase principal no se ajusta del todo a la clase secundaria. En tales casos, volvemos a implementar el método en la clase secundaria.


```python
class Ave:
  def intro(self):
    print("Hay muchos tipos de aves.")
     
  def flight(self):
    print("La mayoria de las aves pueden volar, pero otras no.")
   
class Buho(Ave):
  def flight(self):
    print("Los buhos pueden volar.")
     
class Pinguino(Ave):
  def flight(self):
    print("Los pinguinos no pueden volar.")
     
ave = Ave()
buho = Buho()
pinguino = Pinguino()
 
ave.intro()
ave.flight()
 
buho.intro()
buho.flight()
 
pinguino.intro()
pinguino.flight()
```
